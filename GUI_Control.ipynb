```python
#!/usr/bin/env python3
import rospy
from geometry_msgs.msg import Twist
from std_msgs.msg import Int16, Int16MultiArray
import math
# import cv_signs

# model_path = "/home/rawan/catkin_ws/src/my_robot_package/models/SignModel.pt"
# detector = cv_signs.SignDetector(model_path)

kp_linear = 0
kp_angular = 0
ki_angular = 0
kd_angular = 0
p = 0
i = 0
d = 0
forward_distance = 0
right_distance = 0
left_distance = 0
yaw = 0
max_rotation = 0
min_rotation = 0
rotation_tolerance = 0
linear_speed = 0
rotating_mode = False
last_error = 0
last_letter = ''
seen_letters = []
last_sign = None

def control_1():
    global forward_distance, right_distance, left_distance, yaw, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    
    forward_distance = right_distance = left_distance = yaw = moved = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d

    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        """Keep yaw in [0, 360) degrees"""
        return angle % 360


    def movement():
        global moved
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead

        # sign = cv_signs.get_sign(detector)
        if forward_distance < 8 and moved:
            # if sign == "LEFT":
            #     turn_angle(90, cmd)
            # elif sign == "RIGHT":
            #     turn_angle(-90, cmd)
            # else:
                turn_angle(90, cmd) # turn left for now, untill we have cv readings

        else:
            moved = 1
            cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
            cmd.angular.z = 0
            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            control = pid_control(error)
            control_rad = math.radians(control)
            cmd.angular.z = control_rad
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))
    
    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()

def control_2():
    global forward_distance, right_distance, left_distance, yaw, turned, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    forward_distance = right_distance = left_distance = yaw = turned = moved = target_yaw = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d

    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
                        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        """Keep yaw in [0, 360) degrees"""
        return angle % 360


    def movement():
        global moved, turned
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead
        if forward_distance < 5 and moved:
            turned = 1
            turn_angle(90, cmd) # turn left for now, untill we have cv readings

        else:
            moved = 1
            error = normalize_angle(target_yaw - yaw)
            if abs(error) > rotation_tolerance and turned:
                cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
                error_rad = math.radians(error)
                control = pid_control(error_rad)
                cmd.angular.z = control
            else:
                cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
                cmd.angular.z = 0

            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            error_rad = math.radians(error)
            control = pid_control(error_rad)
            cmd.angular.z = control
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))

    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()

def safe_control():
    forward_distance = right_distance = left_distance = yaw = moved = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d





    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        # Keep yaw in [0, 360) degrees
        return angle % 360


    def movement():
        global moved
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead

        if forward_distance < 8 and moved:
            turn_angle(90, cmd) # turn right

        else:
            moved = 1
            cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
            cmd.angular.z = 0
            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            control = pid_control(error)
            control_rad = math.radians(control)
            cmd.angular.z = control_rad
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))


    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()

```

```python
#!/usr/bin/env python3
import rospy
from geometry_msgs.msg import Twist
from std_msgs.msg import Int16, Int16MultiArray
import math
from tkinter import *   #type: ignore
import cv2
from PIL import Image, ImageTk
import threading
# import cv_signs

# model_path = "/home/rawan/catkin_ws/src/my_robot_package/models/SignModel.pt"
# detector = cv_signs.SignDetector(model_path)

forward_distance = 0
right_distance = 0
left_distance = 0
yaw = 0
moved = False
max_velocity = 0.8
kp_angular = 0
ki_angular = 0
kd_angular = 0
p = 0
i = 0
d = 0
max_rotation = 0
min_rotation = 0
rotation_tolerance = 0
rotating_mode = False
last_error = 0
integral = 0
dt = 0
pub = None
target_yaw = None
last_letter = ''
seen_letters = []
last_sign = None
buttons = []

""" global forward_distance, right_distance, left_distance, yaw, turned, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    """

""" global forward_distance, right_distance, left_distance, yaw, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    """

""" global forward_distance, right_distance, left_distance, yaw, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    """

def control_1():
    global forward_distance, right_distance, left_distance, yaw, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    
    forward_distance = right_distance = left_distance = yaw = moved = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d

    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        """Keep yaw in [0, 360) degrees"""
        return angle % 360


    def movement():
        global moved
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead

        # sign = cv_signs.get_sign(detector)
        if forward_distance < 8 and moved:
            # if sign == "LEFT":
            #     turn_angle(90, cmd)
            # elif sign == "RIGHT":
            #     turn_angle(-90, cmd)
            # else:
                turn_angle(90, cmd) # turn left for now, untill we have cv readings

        else:
            moved = 1
            cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
            cmd.angular.z = 0
            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            control = pid_control(error)
            control_rad = math.radians(control)
            cmd.angular.z = control_rad
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))
    
    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()

def control_2():
    global forward_distance, right_distance, left_distance, yaw, turned, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    forward_distance = right_distance = left_distance = yaw = turned = moved = target_yaw = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d

    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
                        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        """Keep yaw in [0, 360) degrees"""
        return angle % 360


    def movement():
        global moved, turned
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead
        if forward_distance < 5 and moved:
            turned = 1
            turn_angle(90, cmd) # turn left for now, untill we have cv readings

        else:
            moved = 1
            error = normalize_angle(target_yaw - yaw)
            if abs(error) > rotation_tolerance and turned:
                cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
                error_rad = math.radians(error)
                control = pid_control(error_rad)
                cmd.angular.z = control
            else:
                cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
                cmd.angular.z = 0

            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            error_rad = math.radians(error)
            control = pid_control(error_rad)
            cmd.angular.z = control
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))

    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()

def safe_control():
    global forward_distance, right_distance, left_distance, yaw, moved, max_velocity
    global kp_linear, kp_angular, ki_angular, kd_angular, rotation_tolerance
    global last_error, integral, dt, pub, target_yaw
    forward_distance = right_distance = left_distance = yaw = moved = 0
    max_velocity = 0.8
    kp_linear = 1.5

    kp_angular = 0.7
    ki_angular = 0
    kd_angular = 0

    rotation_tolerance = 4

    last_error = 0
    integral = 0
    dt = 0.1

    def pid_control(error):
        global last_error, dt, integral

        # Proportional
        p = kp_angular * error

        # Integral
        integral += error * dt
        i = ki_angular * integral

        # Derivative
        derivative = (error - last_error) / dt if dt > 0 else 0
        d = kd_angular * derivative

        # Save error for next cycle
        last_error = error

        # PID output
        return p + i + d





    def ultrasonics_callback(msg):
        global forward_distance, right_distance, left_distance
        forward_distance = msg.data[0]
        right_distance = msg.data[1]
        left_distance = msg.data[2]
        rospy.loginfo("Forward: %f, Right: %f, Left: %f",
        forward_distance, right_distance, left_distance)

    def yaw_callback(msg):
        global yaw
        yaw = msg.data
        rospy.loginfo("Yaw: %f", yaw)

    def listener():
        rospy.init_node('robot')
        rospy.Subscriber('ultrasonics', Int16MultiArray, ultrasonics_callback)
        rospy.Subscriber('yaw', Int16, yaw_callback)



    def normalize_angle(angle):
        # Wrap angle to [-180, 180] degrees
        return ((angle + 180) % 360) - 180

    def normalize_yaw(angle):
        # Keep yaw in [0, 360) degrees
        return angle % 360


    def movement():
        global moved
        cmd = Twist()

        # safety check
        if forward_distance == 0 or left_distance == 0 or right_distance == 0:
            cmd.linear.x = 0
            cmd.angular.z = 0
            pub.publish(cmd)
            return

        # If obstacle ahead

        if forward_distance < 8 and moved:
            turn_angle(90, cmd) # turn right

        else:
            moved = 1
            cmd.linear.x = min(max_velocity, (forward_distance - 8) * kp_linear)
            cmd.angular.z = 0
            pub.publish(cmd)



    def turn_angle(rotation_angle, cmd):
        # Turn the robot by a relative angle
        global target_yaw, last_error, integral
        start_yaw = yaw
        target_yaw = normalize_yaw(start_yaw + rotation_angle) # try to snap angle here
        turn_rate = rospy.Rate(30) # reduce the turn rate if the imu is very noisy

        last_error = 0
        integral = 0

        error = normalize_angle(target_yaw - yaw)
        while not rospy.is_shutdown() and abs(error) > rotation_tolerance: # increase tolerance if it misses it
            control = pid_control(error)
            control_rad = math.radians(control)
            cmd.angular.z = control_rad
            cmd.linear.x = 0
            pub.publish(cmd)
            turn_rate.sleep()
            error = normalize_angle(target_yaw - yaw)  # update error inside loop

    def snap_angle(angle_deg):
        # Snap an angle in degrees to the closest of [0, 90, -90, 180]
        targets = [0, 90, -90, 180]
        return min(targets, key=lambda x: abs(angle_deg - x))


    def sub_main():
        global pub
        pub = rospy.Publisher('cmd_vel', Twist, queue_size=10)
        listener()
        main_rate = rospy.Rate(10)

        while not rospy.is_shutdown():
            movement()
            main_rate.sleep()

    sub_main()


def gui():
    global kp_angular, ki_angular, kd_angular, p, i, d, max_rotation, min_rotation
    global max_velocity, rotating_mode, last_error, target_yaw
    global forward_distance, right_distance, left_distance, yaw
    global last_letter, seen_letters, last_sign, rotation_tolerance, buttons
    root = Tk()
    root.title("MIA Competition")
    root.state('zoomed')
    canvas = Canvas(root)
    canvas.pack(fill=BOTH, expand=1)

    label = Label(canvas, text="Welcome to MIA Competition GUI", bg="lightblue", font=("Arial", 16))
    label.place(relx=0.5, rely=0.1, anchor=CENTER)

    camera_label = Label(canvas)
    camera_label.place(relx=0.33, rely=0.5, anchor=CENTER)

    text = f"yaw = {yaw}, Rotating: {rotating_mode}, Target: {target_yaw}, Error: {last_error}\n" \
    f"Front Distance = {forward_distance}, Right Distance = {right_distance}, Left Distance = {left_distance}"
    the_keys = f"Kp = {kp_angular}\n\nKi = {ki_angular}\n\nKd = {kd_angular}"
    max_and_min = f"Max Rotation Speed = {max_rotation}\n\nMin Rotation Speed = {min_rotation}\n\nrotation_tolerance = {rotation_tolerance}\n\nLinear Speed = {max_velocity}"
    the_p_i_d = f"P = {p}\n\nI = {i}\n\nD = {d}"
    computer_vision = f"Computer Vision: last letter = {last_letter}, last sign = {last_sign}\nseen letters till now: {",".join(seen_letters)}"

    label_max_min = Label(canvas, text=max_and_min, font=("Arial", 16), justify="left")
    label_max_min.place(relx=0.7, rely=0.45)
    label_text = Label(canvas, text=text, font=("Arial", 16), justify="left")
    label_text.place(relx=0.5, rely=0.35)
    label_keys = Label(canvas, text=the_keys, font=("Arial", 16), justify="left")
    label_keys.place(relx=0.5, rely=0.45)
    label_p_i_d = Label(canvas, text=the_p_i_d, font=("Arial", 16), justify="left")
    label_p_i_d.place(relx=0.65, rely=0.45)
    label_computer_vision = Label(canvas, text=computer_vision, font=("Arial", 16), justify="left")
    label_computer_vision.place(relx=0.5, rely=0.7)

    def action(event):
        global target_to_change, entry, buttons
        if event.keysym == "Return":
            print("Enter pressed")
            num = entry.get()
            try: 
                if num: globals()[target_to_change] = float(num)
                print(f"{target_to_change} changed to {num}")
            except: pass
            entry.destroy()
            for btn in buttons:
                btn.config(state=NORMAL)
            
    def change(x):
        global target_to_change, entry, buttons
        target_to_change = x 
        entry = Entry(canvas, width=10)
        for btn in buttons:
            btn.config(state=DISABLED)
        match x:
            case "kp_angular": entry.place(relx=0.55, rely=0.45)
            case "ki_angular": entry.place(relx=0.55, rely=0.51)
            case "kd_angular": entry.place(relx=0.55, rely=0.57)
            case "max_rotation": entry.place(relx=0.87, rely=0.45)
            case "min_rotation": entry.place(relx=0.87, rely=0.51)
            case "rotation_tolerance": entry.place(relx=0.87, rely=0.57)
            case "max_velocity": entry.place(relx=0.87, rely=0.63)
        entry.bind("<KeyPress>", action)
    
    change_kp_angular = Button(canvas, text="Change", command=lambda x = "kp_angular": change(x))
    change_kp_angular.place(relx=0.6, rely=0.45)
    change_ki_angular = Button(canvas, text="Change", command=lambda x = "ki_angular": change(x))
    change_ki_angular.place(relx=0.6, rely=0.51)
    change_kd_angular = Button(canvas, text="Change", command=lambda x = "kd_angular": change(x))
    change_kd_angular.place(relx=0.6, rely=0.57)
    change_max_rotation = Button(canvas, text="Change", command=lambda x = "max_rotation": change(x))
    change_max_rotation.place(relx=0.92, rely=0.45)
    change_min_rotation = Button(canvas, text="Change", command=lambda x = "min_rotation": change(x))
    change_min_rotation.place(relx=0.92, rely=0.51)
    change_rotation_tolerance = Button(canvas, text="Change", command=lambda x = "rotation_tolerance": change(x))
    change_rotation_tolerance.place(relx=0.92, rely=0.57)
    change_max_velocity = Button(canvas, text="Change", command=lambda x = "max_velocity": change(x))
    change_max_velocity.place(relx=0.92, rely=0.63)

    buttons.extend([change_kp_angular, change_ki_angular, change_kd_angular, change_max_rotation, change_min_rotation, change_rotation_tolerance, change_max_velocity])

    # def camera_frame(): # For CV
        # ret, img = cap.read()
        # if ret:
            # img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            # img = cv2.resize(img, (480, 360))
            # img = Image.fromarray(img)
            # imgtk = ImageTk.PhotoImage(image=img)
            # camera_label.imgtk = imgtk # type: ignore
            # camera_label.config(image=imgtk)
        # camera_label.after(10, camera_frame)

    def on_escape(e):
        root.destroy()

    def update_variables():
        max_and_min = f"Max Rotation Speed = {max_rotation}\n\nMin Rotation Speed = {min_rotation}\n\nrotation_tolerance = {rotation_tolerance}\n\nLinear Speed = {max_velocity}"
        text = f"Front Distance = {forward_distance}, Right Distance = {right_distance}, Left Distance = {left_distance}\n" \
        f"yaw = {yaw}, Rotating: {rotating_mode}, Target: {target_yaw}, Error: {last_error}"
        the_keys = f"Kp = {kp_angular}\n\nKi = {ki_angular}\n\nKd = {kd_angular}"
        the_p_i_d = f"P = {p}\n\nI = {i}\n\nD = {d}"
        computer_vision = f"Computer Vision: last letter = {last_letter}, last sign = {last_sign}\nseen letters till now: {",".join(seen_letters)}"
        label_max_min.config(text=max_and_min)
        label_text.config(text=text)
        label_keys.config(text=the_keys)
        label_p_i_d.config(text=the_p_i_d)
        label_computer_vision.config(text=computer_vision)
        root.after(10, update_variables)

    # camera_frame()
    threading.Thread(target=control_1, daemon=True).start()
    update_variables()
    root.bind("<Escape>", on_escape)
    root.mainloop()

if __name__ == "__main__":
    gui(0, 0, 0, 0)

    # cap.release()
    # cv2.destroyAllWindows()

# roslaunch turtlebot3_gazebo turtlebot3_house.launch
# roslaunch turtlebot3_gazebo turtlebot3_stage_4.launch
# :~/catkin_ws/src/turtlebot3/turtlebot3_description/urdf


# after rotating = check side ultrasonic reading if like old forward one, move slowly with z correction until it is

# check seconds for a full turn, and hard code it using ros.sleep()
```
